```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans
 
X, y= make_blobs(n_samples=210, centers=3, n_features=2, cluster_std=0.5, shuffle=True,random_state=0)
```

```python
plt.scatter(X[:,0], X[:,1], c='red', marker='x')
```

```python
km = KMeans(n_clusters=3, init='random', n_init='auto', max_iter=300,tol=1e-4, random_state=5)

labels = km.fit_predict(X)

plt.figure(figsize=(6,5))
plt.scatter(X[:,0], X[:,1], s=70, c=km.labels_, cmap=plt.cm.prism) 
plt.scatter(km.cluster_centers_[:, 0], km.cluster_centers_[:, 1], marker='*', s=200,
            color='blue', label='Centers')
plt.legend(loc='best')
plt.xlabel('X0')
plt.ylabel('X1')


print(f'Inertia: {km.inertia_}')
```

## Zadanie 2

```python
X_2, y_2 = make_blobs(n_samples=210, centers=3, n_features=2, cluster_std=0.6, shuffle=True,random_state=40)

plt.figure(figsize=(6, 5))
plt.scatter(X_2[:, 0], X_2[:, 1], c='g', marker='.')
plt.title("Wygenerowane dane")
plt.xlabel("X0")
plt.ylabel("X1")
plt.grid(True)
plt.show()
```

```python
km_2 = KMeans(n_clusters=3, init='random', n_init='auto', max_iter=300, tol=1e-4, random_state=40)

labels_2 = km_2.fit_predict(X_2)

plt.figure(figsize=(6,5))
plt.scatter(X_2[:,0], X_2[:,1], s=70, c=km_2.labels_, cmap=plt.cm.prism) 
plt.scatter(km_2.cluster_centers_[:, 0], km_2.cluster_centers_[:, 1], marker='*', s=200,
            color='blue', label='Centers')
plt.legend(loc='best')
plt.xlabel('X0')
plt.ylabel('X1')


print(f'Inertia: {km_2.inertia_}')
```

```python
km_3 = KMeans(n_clusters=5, init='random', n_init='auto', max_iter=300, tol=1e-4, random_state=40)

labels_3 = km_3.fit_predict(X_2)

plt.figure(figsize=(6,5))
plt.scatter(X_2[:,0], X_2[:,1], s=70, c=km_3.labels_, cmap=plt.cm.prism) 
plt.scatter(km_3.cluster_centers_[:, 0], km_3.cluster_centers_[:, 1], marker='*', s=200,
            color='blue', label='Centers')
plt.legend(loc='best')
plt.xlabel('X0')
plt.ylabel('X1')


print(f'Inertia: {km_3.inertia_}')
```

## Zadanie 3

```python
k_means_plus = KMeans(n_clusters=3, init='k-means++', n_init='auto', max_iter=300, tol=1e-4, random_state=40)

labels_plus = k_means_plus.fit_predict(X_2)
print(f'Inertia (k-means++ init): {k_means_plus.inertia_}')

plt.figure(figsize=(6,5))
plt.scatter(X_2[:, 0], X_2[:, 1], s=70, c=labels_plus, cmap=plt.cm.prism, label='Punkty')
plt.scatter(k_means_plus.cluster_centers_[:, 0], k_means_plus.cluster_centers_[:, 1], 
            marker='*', s=200, color='red', label='Centroidy')
plt.title("KMeans++ - Wynik klasteryzacji")
plt.xlabel("X0")
plt.ylabel("X1")
plt.legend(loc='best')
plt.grid(True)
plt.show()
```

## Zadanie 4

```python
X, y = make_blobs(n_samples=300, centers=5, cluster_std=0.8, random_state=43)

k_range = range(1, 11)
inertias = []

for k in k_range:
    km = KMeans(n_clusters=k, init='random', n_init=10, max_iter=300, tol=1e-4, random_state=43)
    km.fit_predict(X)
    inertias.append(km.inertia_)

plt.figure(figsize=(8, 5))
plt.plot(k_range, inertias, 'o-', color='blue')
plt.title('Metoda łokcia – dobór optymalnej liczby klastrów')
plt.xlabel('Liczba klastrów (k)')
plt.ylabel('Wartość zniekształcenia (inertia_)')
plt.xticks(k_range)
plt.grid(True)
plt.show()
```

```python
from sklearn.datasets import make_moons
import seaborn as sns
import pandas as pd

x, y = make_moons(1000, noise=.05, random_state=0)
X_moon = pd.DataFrame(x, columns = ['f1','f2'])
#metoda klasteryzacji
# km -model KMeans
km = KMeans(n_clusters=2, init='k-means++', n_init=10, random_state=0)
#y_km - wynik predykcji
y_km = km.fit_predict(X_moon)
 
#wykres
X_moon['k_means'] = y_km
sns.lmplot(data=X_moon, x='f1', y='f2', fit_reg=False, hue = 'k_means', palette = ['#eb6c6a', '#6aeb6c']).set(title='Algorytm k-srednich')
plt.scatter(km.cluster_centers_[:, 0], km.cluster_centers_[:, 1], c='black', s=100, alpha=0.5)
plt.show()
```
