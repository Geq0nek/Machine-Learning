```python
import pandas as pd 
import csv
import math
import matplotlib.pyplot as plt
import numpy as np
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression
```

```python
#Wczytanie danych uczÄ…cych 
with open('./lin_reg_training.csv', 'r') as f:
    csvreader = csv.reader(f, delimiter=',')
    data = [row for row in csvreader]
    data = np.array(data, dtype=np.float32)
```

```python
x_train = data[:, 0]
y_train = data[:, 1]
plt.scatter(x_train, y_train)
plt.xlabel('x')
plt.ylabel('y')
```

```python
standard_scaler = StandardScaler()
x_train = x_train.reshape(-1,1)

x_train = standard_scaler.fit_transform(x_train)

poly_features = PolynomialFeatures(degree=6)
x_train_poly = poly_features.fit_transform(x_train)
```

```python
#Dane testowe w celu sprawdzenia efketywnoÅ›ci modelu
with open('./lin_reg_test.csv', 'r') as f:
    csvreader = csv.reader(f, delimiter=',')
    data = [row for row in csvreader]
    data = np.array(data, dtype=np.float32)

x_test = data[:, 0]
y_test = data[:, 1]
x_test = x_test.reshape(-1,1)
x_test = standard_scaler.transform(x_test)
poly_features = PolynomialFeatures(degree=6)
x_train_poly = poly_features.fit_transform(x_train)
x_test_poly = poly_features.transform(x_test)
model = LinearRegression()
model.fit(x_train_poly, y_train)
```

```python
x_range = np.linspace(x_train.min(), x_train.max(), 100).reshape(-1, 1)
x_range_poly = poly_features.transform(x_range)

y_range_pred = model.predict(x_range_poly)
y_test_pred = model.predict(x_test_poly)
```

```python
#Wyrysowanie wykresu
plt.figure(figsize=(8, 6))
plt.scatter(x_train, y_train, c='blue', label='Training data')
plt.scatter(x_test, y_test, c='orange', label='Test data')
plt.plot(x_range, y_range_pred, color='blue', label='Polynomial fit')
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Polynomial Regression Fit')
plt.legend()
plt.show()
```

```python
#Wyliczenie kosztu dla danych testowych
cost = np.mean((y_test_pred - y_test) ** 2) / 2
print(f"Cost on test set: {cost}")

#Wyliczenie kosztu dla danych uczÄ…cych
y_train_pred = model.predict(x_train_poly)
train_cost = np.mean((y_train_pred - y_train) ** 2) / 2
print(f"Cost on training set: {train_cost}")
```

```python

def h_fun(X, theta):
    '''
    :param X: ndarray postaci (n+1, m).
    :param theta: macierz parametrÃ³w do optymalizacji postaci (n+1, 1)
    :return: 
    '''
    return np.dot(theta.T, X) # przykÅ‚adowa realizacja, @ - mnoÅ¼enie macierzowe

def calculate_cost(X, y, theta, lambda_=0):
    m = X.shape[1]
    h = h_fun(X, theta)
    lin_reg_cost = np.mean((h - y) ** 2) / 2
    sum_of_thetas = (theta[1:, :] ** 2).sum()
    return lin_reg_cost + sum_of_thetas * lambda_ / (2 * m)

def compute_gradients(X, y, theta, lambda_=0):
    m = X.shape[1]
    h = h_fun(X, theta) 
    gradients = (X @ (h - y).T) / m 
    theta_reg = np.concatenate([np.zeros((1, 1)), theta[1:, :]], axis=0)
    gradients += (lambda_ / m) * theta_reg
    return gradients
```

```python
def train_model(x_poly, y, theta_init, alpha, iterations, lambda_):
    theta = theta_init.copy()
    for _ in range(iterations):
        gradients = compute_gradients(x_poly.T, y, theta, lambda_)
        theta -= alpha * gradients
    return theta

lambdas = [10, 1, 0.01, 0.001, 0.0001]
theta_init = np.zeros((x_train_poly.shape[1], 1))

alpha = 0.01
iterations = 100_000
results = {}

for lambda_ in lambdas:
    theta_opt = train_model(x_train_poly, y_train, theta_init, alpha, iterations, lambda_)
    
    train_cost = calculate_cost(x_train_poly.T, y_train, theta_opt, lambda_=0)
    test_cost = calculate_cost(x_test_poly.T, y_test, theta_opt, lambda_=0)
    
    results[lambda_] = {
        "theta": theta_opt,
        "train_cost": train_cost,
        "test_cost": test_cost
    }
    
    print(f"Lambda: {lambda_}, Train Cost: {train_cost:.4f}, Test Cost: {test_cost:.4f}")
```

```python
# Tworzenie nowej siatki wykresÃ³w
fig, axes = plt.subplots(2, 3, figsize=(18, 10))
axes = axes.ravel()

for i, lambda_ in enumerate(lambdas):
    theta_opt = results[lambda_]["theta"]
    y_range_pred = x_range_poly @ theta_opt

    ax = axes[i]
    ax.scatter(x_train, y_train, c='blue', label='Training data', alpha=0.6)
    ax.scatter(x_test, y_test, c='orange', label='Test data', alpha=0.3)
    ax.plot(x_range, y_range_pred, color='blue', label='Polynomial fit')
    ax.set_title(f'Lambda = {lambda_}')
    ax.set_xlabel('X')
    ax.set_ylabel('Y')
    ax.grid(True)
    ax.legend()

# UsuniÄ™cie pustego wykresu jeÅ›li lambdas < liczba subplotÃ³w
if len(lambdas) < len(axes):
    for j in range(len(lambdas), len(axes)):
        fig.delaxes(axes[j])

plt.tight_layout()
plt.show()
```

# Regresja logistyczna

```python
def sigmoid(z):
    return 1.0 / (1.0 + np.exp(-z))

def h_fun(X, theta):
    z = np.dot(X, theta)
    return sigmoid(z)

def calculate_cost(X, y, theta, lambda_=0):
    m = X.shape[0] 
    h = h_fun(X, theta)
    epsilon = 1e-10  
    
    cost = (-y * np.log(np.clip(h, epsilon, 1 - epsilon)) - (1 - y) * np.log(np.clip(1 - h, epsilon, 1 - epsilon))).mean()

    reg_cost = (lambda_ / (2 * m)) * np.sum(np.square(theta[1:])) 
    
    return cost + reg_cost

def compute_gradients(X, y, theta, lambda_=0):
    m = X.shape[0]  
    h = h_fun(X, theta)
    gradients = (X.T @ (h - y)) / m  
    
    gradients[1:] += (lambda_ / m) * theta[1:]

    return gradients


def train_logistic_model(X, y, theta_init, alpha, iterations, lambda_):
    theta = theta_init.copy()
    for _ in range(iterations):
        grad = compute_gradients(X.T, y, theta, lambda_)
        theta -= alpha * grad
    return theta
```

```python
# Parametry treningu
theta_init = np.zeros((x_train_poly.shape[1], 1))
iterations = 10000
alpha = 0.1
lambdas_log = [0, 10, 1, 0.5, 0.1, 0.05, 0.01]

# --- Trening modeli logistycznych ---
log_results = {}
for lambda_ in lambdas_log:
    theta_opt = train_logistic_model(x_train_poly, y_train, theta_init, alpha, iterations, lambda_)
    train_cost = calculate_cost(x_train_poly, y_train, theta_opt, lambda_=0)
    test_cost = calculate_cost(x_train_poly, y_train, theta_opt, lambda_=0)
    log_results[lambda_] = {
        "theta": theta_opt,
        "train_cost": train_cost,
        "test_cost": test_cost
    }

# --- Tabela wynikÃ³w ---
log_cost_df = pd.DataFrame([
    {
        "lambda": lam,
        "train_cost": log_results[lam]["train_cost"],
        "test_cost": log_results[lam]["test_cost"]
    }
    for lam in lambdas_log
])


fig, axes = plt.subplots(2, 4, figsize=(20, 10))
axes = axes.ravel()

for i, lambda_ in enumerate(lambdas_log):
    theta_opt = log_results[lambda_]["theta"]
    y_pred_prob = sigmoid(x_range_poly @ theta_opt)

    ax = axes[i]
    ax.scatter(x_train, y_train, c='orange', label='Dane binarne', alpha=0.6)
    ax.plot(x_range, y_pred_prob, color='blue', label='Predykcja (sigmoid)')
    ax.set_title(f"Lambda = {lambda_}")
    ax.set_xlabel("X")
    ax.set_ylabel("P(class = 1)")
    ax.set_ylim(-0.05, 1.05)
    ax.grid(True)
    ax.legend()

# UsuniÄ™cie pustych subplotÃ³w jeÅ›li sÄ…
for j in range(len(lambdas_log), len(axes)):
    fig.delaxes(axes[j])

plt.suptitle("Regresja logistyczna - dopasowanie modelu dla rÃ³Å¼nych lambda", fontsize=16)
plt.tight_layout(rect=[0, 0.03, 1, 0.95])
plt.show()

# ZwrÃ³cenie wynikÃ³w
log_cost_df
```
